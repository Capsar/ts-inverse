```python
import torch
from torch import nn, optim

import pandas as pd
import datetime
from ts_inverse.datahandler import ElectricityDataSet
import matplotlib.pyplot as plt
import numpy as np


def plot_inputs_targets(input_sequence, target_sequence, restored_input_sequence, restored_target_sequence):
    fig, axes = plt.subplots(input_sequence.shape[0], 1, figsize=(16, 3 * input_sequence.shape[0]))
    axes = np.atleast_1d(axes)  # Ensure axes is always an array for consistency
    inputs_sequence_length = input_sequence.shape[1]
    targets_sequence_length = target_sequence.shape[1]

    for i, ax in enumerate(axes):
        original_x_axis = np.arange(inputs_sequence_length + targets_sequence_length)
        if input_sequence is not None:
            ax.plot(original_x_axis[:inputs_sequence_length], input_sequence[i].detach().cpu().numpy(), label="Original Inputs")
        if restored_input_sequence is not None:
            ax.plot(original_x_axis[:inputs_sequence_length], restored_input_sequence[i].detach().cpu().numpy(), label="Restored Inputs", linestyle=":")

        if target_sequence is not None:
            ax.plot(original_x_axis[-targets_sequence_length:], target_sequence[i].detach().cpu().numpy(), label="Original Targets")
        if restored_target_sequence is not None:
            ax.plot(original_x_axis[-targets_sequence_length:], restored_target_sequence[i].detach().cpu().numpy(), label="Restored Targets", linestyle=":")
        ax.legend()
    plt.show()


df_elec = pd.read_csv("../data/Electricity370/LD2011_2014_first_40_consumers.csv", index_col="Time", parse_dates=["Time"])
df_elec.head()

series = df_elec.loc[:, df_elec.columns[0]]

el_dataset = ElectricityDataSet(
    series, step_size=datetime.timedelta(hours=24), observation_length=datetime.timedelta(days=1), target_length=datetime.timedelta(days=1)
)

# Generate synthetic data
torch.manual_seed(0)


# Define the linear model
class LinearNet(nn.Module):
    def __init__(self, input_size, output_size, hidden_size=64):
        super(LinearNet, self).__init__()
        self.act = nn.Sigmoid()
        self.fc_1 = nn.Linear(input_size, hidden_size)
        self.fc_2 = nn.Linear(hidden_size, hidden_size)
        self.fc_head = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.act(self.fc_1(x))
        x = self.act(self.fc_2(x))
        return self.fc_head(x)


# Initialize the network and optimizer
input_length = 96
output_length = 96

net = LinearNet(input_size=input_length, output_size=output_length)
criterion = nn.MSELoss()
net_optimizer = optim.SGD(net.parameters(), lr=0.01)

offset = 6
batch_size = 1
training_steps = 10
use_one_shot_target_reconstruction = True
reconstruction_steps = 5000
for batch in range(training_steps):
    inputs, targets = el_dataset[offset + batch : offset + batch + batch_size]
    inputs, targets = inputs[:, :, 0], targets[:, 0:output_length, 0]

    plot_inputs_targets(inputs, targets, inputs, targets)

    # Prepare data
    actual_inputs = inputs
    actual_targets = targets

    # Training process
    net_optimizer.zero_grad()
    predicted_targets = net(actual_inputs)
    loss = criterion(predicted_targets, actual_targets)
    original_dy_dx = torch.autograd.grad(loss, net.parameters(), create_graph=True)

    with torch.no_grad():
        # Extract gradients of the last layer
        grad_w = original_dy_dx[-2]  # Shape: [output_size, hidden_size]
        grad_b = original_dy_dx[-1]  # Shape: [output_size]

        # Since gradients are averaged, multiply by batch_size to get the sum of gradients
        grad_w = grad_w * batch_size  # Shape: [output_size, hidden_size]
        grad_b = grad_b * batch_size  # Shape: [output_size]

        # Ensure the dimensions of grad_b are suitable for matrix operations
        grad_b = grad_b.unsqueeze(-1)  # Shape: [output_size, 1]

        # Use pseudo-inverse for the reconstruction
        pseudo_inverse_grad_b = torch.pinverse(grad_b)  # Shape: [1, output_size]
        reconstructed_head_inputs = torch.mm(pseudo_inverse_grad_b, grad_w)  # Shape: [1, hidden_size]

        # Expand the reconstructed inputs to match the batch size
        reconstructed_head_inputs = reconstructed_head_inputs.expand(batch_size, -1)

        # Reconstruct the targets
        reconstructed_targets = net.fc_head(reconstructed_head_inputs) - (grad_b.T * (output_length / 2))
        reconstructed_targets = reconstructed_targets.squeeze(-1)  # Shape: [batch_size, output_size]
        plot_inputs_targets(actual_inputs, actual_targets, None, reconstructed_targets)

    reconstructed_inputs = torch.rand_like(actual_inputs).requires_grad_(True)
    if use_one_shot_target_reconstruction:
        reconstructed_targets.requires_grad_(True)
    else:
        reconstructed_targets = torch.rand_like(actual_targets).requires_grad_(True)
    optimizer = optim.Adam([reconstructed_inputs, reconstructed_targets], lr=0.01)
    for j in range(reconstruction_steps):
        optimizer.zero_grad()

        net.zero_grad()
        predicted_targets = net(reconstructed_inputs)
        reconstruct_loss = criterion(predicted_targets, reconstructed_targets)
        dummy_dy_dx = torch.autograd.grad(reconstruct_loss, net.parameters(), create_graph=True)

        dy_dx_loss = sum((d_g - o_g.clone()).abs().sum() for d_g, o_g in zip(dummy_dy_dx, original_dy_dx))
        dy_dx_loss.backward(retain_graph=True)
        optimizer.step()
        if j % int(reconstruction_steps / 10) == 0:
            print(f"Step {j}/{reconstruction_steps}: {reconstruct_loss}")

    plot_inputs_targets(actual_inputs, actual_targets, reconstructed_inputs, reconstructed_targets)

    # print('Predicted Targets:', reconstructed_targets.detach())

    loss.backward()
    net_optimizer.step()
    print("NEXT TRAINING SAMPLES STEP")
```
